```python
sorted_freq_rank = word_info["freq_rank"].sort_values()
sorted_freq_rank = sorted_freq_rank.values
```

```python
min_freq_rank = min(sorted_freq_rank)
max_freq_rank = max(sorted_freq_rank)
freq_rank_set = set(sorted_freq_rank)

absent_values = []

for i in range(min_freq_rank, max_freq_rank):
    if not i in freq_rank_set:
        absent_values.append(i)

print(absent_values)
```

```python
# defs: [[{def: _, examples: _}, {}], []] -> [][]
# [{def: def1, example: _}, {}] -> [def1, def2, def3, ...]
def extract_definitions(definition_list):
    def extract_definition(definition):
        return definition["def"]
    
    definition_list = list(map(extract_definition, definition_list))
    print(definition_list)
    return definition_list

print(list(word_info["defs"]))

list(map(extract_definitions, list(word_info["defs"])))
```

```python
definition_series = word_info["defs"]
for definition_entry in definition_series:
    if definition_entry == None:
        print("None encountered")
        continue

    new_definition_entry = []
    for definition in definition_entry:
        new_definition_entry.append(definition["def"])
```
